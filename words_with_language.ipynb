```python
import pandas as pd
import os
```

```python
path = r'C:\Users\HP\Downloads\most_common_words\wordss'
```

```python
d = {'word' : [], 'language' : []}
```

```python
words = []
lang = []
for f in os.listdir(path):
    with open(os.path.join(path,f),encoding='utf8') as file:
        string = file.read()
        lst = string.split('\n')
        l = len(lst)
        for i in range(1,l):
            string1 = lst[i]
            lst2 = string1.split(' ')
            words.append(lst2[1])
            lang.append(f[:-10])
```

```python
d['word'] = words
d['language'] = lang
```

```python
print(d)
```

```python
df = pd.DataFrame.from_dict(d)
```

```python
df_final = df.sample(frac = 1)
df_final.to_csv('final_result.csv', index = False)
file1 = open('final_result.csv','r',encoding = 'utf8')
l = file1.readlines()
file1.close()
f2 = open('shuffled_final_result1.csv','w', encoding = 'utf8')
f2.writelines(l)
f2.close()
```
